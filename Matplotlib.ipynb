# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
```

## Line charts

```python
#to create a line chart
#first one is x axis, which is horizontal numbers and here is year
plt.plot(year, pop)
#to label axis
plt.xlabel('Year')
plt.ylabel('Population')
#to add a title
plt.title('World Population Projects')
#to change the Y (vertical) axis to start at 0 and change Y axis tick label names
plt.yticks([0, 2, 4, 6], ['0', '2 Billion', '4 Billion', '6 Billion'])
plt.show()
```

```python
#to add historical data
year = [1800, 1850 ,1900] + year
pop = [1.0, 1.262, 1.650] + pop

plt.plot(year, pop)
#to label axis
plt.xlabel('Year')
plt.ylabel('Population')
#to add a title
plt.title('World Population Projects')
#to change the Y (vertical) axis to start at 0 and change Y axis tick label names
plt.yticks([0, 2, 4, 6], ['0', '2 Billion', '4 Billion', '6 Billion'])
plt.show()
```

## Scatterplots

```python
#to create a scatterplot
#x axis is first and is on the horizontal, here it's year
plt.scatter(year,pop)
plt.grid(True)
plt.show()
```

```python
#to select the last thing in a list
print(year[-1])
```
